Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
  >Answer:
  1. Type 1 error also called False positive occurs when the null hypothesis is incorrectly rejected, meaning we conclude there is an effect or difference when in reality there is none. This can lead to unnecessary actions, such as adopting ineffective treatments or methods.
  2. Type 2 error called False negative occurs when the null hypothesis is not rejected  despite an actual effect or difference existing. This means failing to detect something significant and may result in missed opportunities or unresolved issues.
  3. Both errors impact decision making by introducing risks. Type 1 errors can cause unwarranted changes, while Type 2 errors can prevent important interventions.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
  >Answer:
  1. The P - value in hypothesis testing measures the probability of observing the data or something more extreme if the null hypothesis is true.
  2. A small P value(=<0.05)indicates strong evidence against the null hypothesis, suggesting it should be rejected. Conversely, a P value greater than 0.05 suggests the data doesnot provide sufficient evidence to reject the null hypothesis.

```python
# Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
import numpy as np
from scipy.stats import norm

#Example data
data = [88,92,94,96,97,100,101,103,105,110]
population_mean = 100
population_std = 15

sample_mean = np.mean(data)
n = len(data)

z = (sample_mean - population_mean) / (population_std - np.sqrt(n))
p_value = 2 * (1- norm.cdf(abs(z)))

print("Z - score:", z)
print ("P -value:", p_value)
```

```python
# Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

non_normal = np.random.exponential(scale=2, size=10000)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(non_normal, bins=30, color= "skyblue", edgecolor= "black")
plt.title("Original non_normal Distribution")

sample_means = [np.mean(np.random.choice(non_normal, size=30)) for _ in range(10000)]

plt.subplot(1,2, 2)
plt.hist(sample_means, bins=30, color="orange", edgecolor="black")
plt.title("Distribution of Sample Mean (CLT)")

plt.tight_layout()
plt.show()
```

```python
# Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence= 0.95):
  n = len(data)
  mean = np.mean(data)
  sem = stats.sem(data)
  margin = sem * stats.t.ppf((1 + confidence) / 2, n-1)
  lower = mean - margin
  upper = mean + margin

  plt.figure(figsize= (6, 3))
  plt.errorbar(1, mean, yerr= margin, fmt= "o", capsize= 5)
  plt.xlim(0, 2)
  plt.xticks([1], ["Sample Mean"])
  plt.title(f"{int(confidence*100)} % Confidence Interval")
  plt.ylabel("Value")
  plt.show()

  print("Mean", mean)
  print(f"{int(confidence*100)} % Confidence Interval: [{lower: .2f}, {upper: .2f}]")

data = [8,9,10,8,9,9,10,8,9,10]
plot_confidence_interval(data)
```

```python
# Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
import numpy as np
from scipy.stats import chisquare

observed = [20,30,50]
expected = [33.3,33.3,33.4]

chi2_stat, p_value = chisquare(f_obs= observed, f_exp= expected)

print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)
```
